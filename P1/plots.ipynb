```python
times = {
    10: [0.000049, 0.000098, 0.000164, 0.000194, 0.000257, 0.000329, 0.000390, 0.000498],
    100: [0.003562, 0.002945, 0.003077, 0.003679, 0.003762, 0.004316, 0.006526, 0.006742],
    500: [0.122906, 0.098454, 0.095256, 0.078726, 0.093228, 0.085395, 0.089147, 0.139970],
    1000: [0.706596, 0.494983, 0.409568, 0.380961, 0.438014, 0.696560, 0.699248, 0.814681],
    2000: [4.771281, 2.944120, 2.489187, 2.250240, 2.738699, 2.887806, 3.566102, 3.935964],
    3000: [15.563150, 9.321857, 7.315904, 7.155480, 9.122288, 10.290178, 10.620403, 13.004947]
}
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
speedup = {}
for n in times.keys():
    l = []
    for t in times[n]:
        l.append(times[n][0] / t)
    speedup[n] = l
```

```python
n_list = list(times.keys())
for p in range(8):
    print(f'{p+1}', end=' & ')
    for n in n_list:
        print(f'{times[n][p]:.6f}', end=' & ')
    print('\b\b\\\\')
```

```python
p = list(range(1, 9))
plt.figure(figsize=(10, 5))
for n in times.keys():
    plt.plot(p, times[n], label="n=" + str(n))
plt.xlabel("Processors")
plt.ylabel("Speedup")
plt.legend()
plt.savefig("speedup.png")
```

```python
eff = {}
for n in speedup.keys():
    l = []
    for p, s in enumerate(speedup[n]):
        l.append(s/(p+1))
    eff[n] = l
```

```python
p = list(range(1, 9))
plt.figure(figsize=(10, 5))
for n in eff.keys():
    plt.plot(p, eff[n], label="n=" + str(n))
plt.xlabel("Processors")
plt.ylabel("Efficiency")
plt.legend()
plt.savefig("eff.png")
```

```python
cost = {}
for n in times.keys():
    l = []
    for p, t in enumerate(times[n]):
        l.append((p+1)*t)
    cost[n] = l
```

```python
p = list(range(1, 9))
plt.figure(figsize=(10, 5))
for n in cost.keys():
    plt.plot(p, cost[n], label="n=" + str(n))
plt.xlabel("Processors")
plt.ylabel("Cost")
plt.legend()
plt.savefig("cost.png")
```
