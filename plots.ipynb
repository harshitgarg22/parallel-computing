```python
def getsf(s, p):
    return (1/s - 1/p) / (1 - 1/p)
```

```python
sfraction = {}
for n in speedup.keys():
    l = []
    for p, s in enumerate(speedup[n]):
        if p == 0:
            l.append(1)
        else:
            l.append(max(0, getsf(s, p+1)))
    sfraction[n] = l
```
